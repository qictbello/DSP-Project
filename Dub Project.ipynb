```python
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
import moviepy.editor as mp

# Define the input video path
input_video_path = "/content/drive/MyDrive/dsp/Dub Project/One Piece Episode 5.mp4"

# Define the output video path (split video) and audio path
output_video_path = "/content/drive/MyDrive/dsp/Dub Project/splitvideo.mp4"
output_audio_path = "/content/drive/MyDrive/dsp/Dub Project/splitaudio.wav"

# Define the start and end times for the split (in seconds)
start_time = 17 * 60 + 15  # 17:15 in seconds
end_time = 20 * 60 + 45  # 20:45 in seconds

# Split the video
ffmpeg_extract_subclip(input_video_path, start_time, end_time, targetname=output_video_path)

# Extract the audio from the split video
video_clip = mp.VideoFileClip(output_video_path)
audio_clip = video_clip.audio
audio_clip.write_audiofile(output_audio_path)

# Close the video clip
video_clip.close()
```

# Modify Voice Buggy - Bello

```python
!pip install scipy numpy
```

```python
from scipy.io import wavfile
from scipy.ndimage import interpolation
import numpy as np

for i in range(1, 9):
    # Input Path
    input_wav_path = f'/content/drive/MyDrive/dsp/Dub Project/Recording/Buggy/rec{i}.wav'
    # Load
    sample_rate, audio_data = wavfile.read(input_wav_path)
    #(0.5 for higher pitch, 2.0 for lower pitch)
    speed_factor = 0.85
    # Use interpolation
    modified_audio = interpolation.zoom(audio_data, speed_factor)
    # OUTPUT
    output_wav_path = f'/content/drive/MyDrive/dsp/Dub Project/Modified/Buggy/rec{i}edit.wav'
    modified_audio = (modified_audio / np.max(np.abs(modified_audio)) * 32767).astype(np.int16)

    # Save
    wavfile.write(output_wav_path, sample_rate, modified_audio)
```

# Modify Voice Ross

```python
import numpy as np
import scipy.signal as signal
import librosa
import soundfile as sf

# Load the audio file
audio_file, sr = librosa.load('/content/drive/MyDrive/Dub Project/Recording/Zoro/Zoro 1.wav', sr=None)

shifted_audio = librosa.effects.pitch_shift(audio_file, n_steps=-3, sr=sr)  # Adjust the value as needed

# Adjust the cutoff frequency as needed for filtering
cutoff_hz = 6000
nyquist = 0.7 * sr
normal_cutoff = cutoff_hz / nyquist

# Design a low-pass Butterworth filter
b, a = signal.butter(10, normal_cutoff, btype='low', analog=False)

# Apply the filter to the audio
filtered_audio = signal.filtfilt(b, a, shifted_audio)

# Save the processed audio as a WAV file
sf.write('/content/drive/MyDrive/Dub Project/Modified/Zoro/Zoro 1.wav', filtered_audio, sr)
```

```python
## import record to original wav file
from google.colab import drive
from google.colab import files
from pydub import AudioSegment

#main_wav_path = '/content/drive/MyDrive/dsp/Dub Project/removedmusicsplitaudio.wav'
main_wav_path = '/content/drive/MyDrive/dsp/Dub Project/Modified/Video/output.wav'
your_wav_path = '/content/drive/MyDrive/dsp/Dub Project/Modified/Zoro/8.wav'
output_wav_path = '/content/drive/MyDrive/dsp/Dub Project/Modified/Video/output.wav'

main_audio = AudioSegment.from_file(main_wav_path, format="wav")
your_audio = AudioSegment.from_file(your_wav_path, format="wav")

start_timestamp = 3 * 60 * 1000 + 4 * 1000
end_timestamp = 3 * 60 * 1000 + 7 * 1000

main_audio_before = main_audio[:start_timestamp]
main_audio_middle = main_audio[start_timestamp:end_timestamp]
main_audio_after = main_audio[end_timestamp:]

mixed_audio = main_audio_middle.overlay(your_audio)

combined_audio = main_audio_before + mixed_audio + main_audio_after

combined_audio.export(output_wav_path, format="wav")
files.download(output_wav_path)
```
